```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, dense_rank
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DateType, FloatType
```

```python
spark = (
    SparkSession.builder
      .master('local')
      .appName('Projeto_20211108')
      .config('spark.ui.port', '4050')
      .getOrCreate()
)
```

```python
# Criando o Dataframe Spark a partir do Dataframe do Pandas

# No esquema, alguns nomes de coluna serão alterados

esquema = (
    StructType()
      .add("ID",IntegerType(),True)
      .add("Ano_Nascimento",IntegerType(),True)
      .add("Escolaridade",StringType(),True)
      .add("Estado_Civil",StringType(),True)
      .add("Salario_anual",FloatType(),True)  # Nome de coluna alterado
      .add("Criancas_residentes",IntegerType(),True)
      .add("Adolesc_residentes",IntegerType(),True)
      .add("Data_cliente",DateType(),True)
      .add("Recencia",IntegerType(),True)  # Nome de coluna alterado
      .add("QtdVinhos",IntegerType(),True)  # Nome de coluna alterado
      .add("QtdFrutas",IntegerType(),True)  # Nome de coluna alterado
      .add("QtdProdutosCarne",IntegerType(),True)  # Nome de coluna alterado
      .add("QtdProdutosPeixe",IntegerType(),True)  # Nome de coluna alterado
      .add("QtdProdutosDoces",IntegerType(),True)  # Nome de coluna alterado
      .add("QtdProdutosOuro",IntegerType(),True)  # Nome de coluna alterado
      .add("NumComprasPromo",IntegerType(),True)
      .add("NumComprasWeb",IntegerType(),True)  # Nome de coluna alterado
      .add("NumComprasCatalogo",IntegerType(),True)  # Nome de coluna alterado
      .add("NumComprasLoja",IntegerType(),True)
      .add("NumVisitasWebMes",IntegerType(),True)
      .add("Oferta_Aceita3",IntegerType(),True)  # Nome de coluna alterado
      .add("Oferta_Aceita4",IntegerType(),True)  # Nome de coluna alterado
      .add("Oferta_Aceita5",IntegerType(),True)  # Nome de coluna alterado
      .add("Oferta_Aceita1",IntegerType(),True) # Nome de coluna alterado
      .add("Oferta_Aceita2",IntegerType(),True)  # Nome de coluna alterado
      .add("Reclamacao",IntegerType(),True)  # Nome de coluna alterado
      .add("Resposta",IntegerType(),True)  # Nome de coluna alterado
)

sparkDF = spark.createDataFrame(df, schema=esquema)
```

```python
sparkDF.show(5)
```

```python
# Imprimir o esquema para ver se o StructType funcionou

sparkDF.printSchema()
```

```python
# Convertendo o tipo dos dados na coluna "Salario_anual" para integer

sparkDF2 = sparkDF.withColumn('Salario_anual', sparkDF['Salario_anual'].cast("integer"))

# Atribui o comando para sparkDF2, para ficar com o sparkDF como backup
```

```python
sparkDF2.printSchema()
```

```python
# Criando uma nova coluna para somar a quantidade total de produtos comprados por usuário

sparkDF2 = (
    sparkDF2.withColumn('QtdProdutosTotal', 
                        sparkDF2['QtdVinhos'] + sparkDF2['QtdFrutas'] + sparkDF2['QtdProdutosCarne'] + 
                        sparkDF2['QtdProdutosPeixe'] + sparkDF2['QtdProdutosDoces'] + sparkDF2['QtdProdutosOuro'])
)
```

```python
sparkDF2.show(5)
```

```python
# Calcula a média do ano de nascimento

sparkDF2.select(F.avg('Ano_Nascimento')).show()
```

```python
# Soma salario anual total

sparkDF2.select(F.sum('Salario_Anual')).show()
```

```python
sparkDF2.count()
```

```python
# Calcula média de salario anual dos clientes da loja
salario_medio = 115599555 / 2237
salario_medio
```

```python
# Cria uma nova coluna para informar se o salário do cliente é acima da média, usando valores do tipo bool

sparkDF2 = (
    sparkDF2.withColumn('SalarioAcimaMedia', F.when(sparkDF2['Salario_anual'] > salario_medio, True).otherwise(False))
)
```

```python
sparkDF2.show(5)
```

```python
# Ordena o DF por número de Compras na Web, para descobrir os clientes que mais compram por esse canal

sparkDF2.orderBy(F.col('NumComprasWeb').desc()).show(10)
```

```python
# Ordena o DF por número de Compras na Loja, selecionando somente estado civil e salario anual

sparkDF2.select('ID', 'Estado_Civil', 'Salario_anual', 'NumComprasLoja').orderBy(F.col('NumComprasLoja').desc()).show(20)
```

```python
# Calcula quantidade de clientes por Escolaridade

sparkDF2.groupBy(F.col('Escolaridade')).count().show()
```

```python
# Calcula a média de produtos totais comprados por escolaridade

sparkDF2.groupBy(F.col('Escolaridade')).mean('QtdProdutosTotal').show()
```

```python
# Calcula a soma e média de salários e soma e média de produtos agrupando por Estado civil

(
sparkDF2.groupBy(F.col('Estado_Civil')).agg(
    F.sum('Salario_Anual').alias('SomaSalario'),
    F.avg('Salario_Anual').alias('MediaSalario'),
    F.sum('QtdProdutosTotal').alias('SomaProdutosTotal'),
    F.avg('QtdProdutosTotal').alias('MediaProdutosTotal')).show()
)
```

```python
sparkDF2.groupBy(F.col('Estado_Civil')).count().show()

'''
Por algum motivo que desconheço, as linhas com valor <NA> na coluna Estado_Civil 
foram importadas do Pandas como: 
'net.razorvine.pickle.objects.ClassDictConstructor@26e8046d'

Se eu tento alterar esse valor com o comando abaixo, todos os valores na coluna Estado_Civil ficam com null...

sparkTeste = sparkDF2.withColumn('Estado_Civil', F.when(F.col('Estado_Civil') == 'net.razorvine.pickle.objects.ClassDictConstructor@26e8046d', '<NA>'))
'''
```

```python
# Soma as quantidades totais de cada produto
# Assim sabemos quais produtos vendem mais ou menos

(
sparkDF2.select(F.sum('QtdVinhos'), F.sum('QtdFrutas'), 
                F.sum('QtdProdutosCarne'), F.sum('QtdProdutosPeixe'), 
                F.sum('QtdProdutosDoces'), F.sum('QtdProdutosOuro')).show()
)
```

```python
# Cria a coluna "num_linha" usando a função row_numer e window function com orderBy, do mais velho ao mais novo cliente

sparkDF2.withColumn('num_linha', F.row_number().over(Window.orderBy('Ano_Nascimento'))).show(30)
```

```python
# Cria um ranking do cliente que mais comprou ao que menos comprou

w0 = Window.orderBy(F.col('QtdProdutosTotal').desc())
sparkDF2.withColumn('ranking_compras', F.dense_rank().over(w0)).show(100)
```

```python
# Salvando o DF Spark no Drive como arquivo csv

(
    sparkDF2.write.format("csv")
    .option("header", "true")
    .option("inferschema", "true")
    .option("delimiter", ",")
    .save("/content/drive/MyDrive/Market_Campaing_Spark.csv")
)
```

# Spark SQL

```python
sparkDF2.createOrReplaceTempView('marketing')
```

```python
# Testando o spark.sql

spark.sql('''

SELECT
  *
FROM
  marketing
LIMIT
  20

'''
).show()
```

```python
# Mostra todos os graus de escolaridade distintos

spark.sql(
'''
SELECT 
  DISTINCT Escolaridade
FROM
  marketing
'''
).show()
```

```python
# Calcula o salário anual médio

spark.sql(
'''
SELECT
  AVG(Salario_Anual)
FROM
  marketing
'''
).show()
```

```python
# Calcula o salário médio das pessoas com PhD

spark.sql(
'''
SELECT
  AVG(Salario_Anual) AS MediaSalarioPhd
FROM
  marketing
WHERE
  Escolaridade = 'PhD'
'''
).show()
```

```python
# Mostra somente as pessoas com salario anual acima do salario medio (calculado acima)

spark.sql(
'''
SELECT
  ID,
  Escolaridade,
  Salario_anual,
  QtdProdutosTotal
FROM
  marketing
WHERE
  Salario_anual > 51676

'''
).show(100)
```

```python
# Seleciona somente os Graduados que compraram mais de 1500 produtos no total

spark.sql(
'''
SELECT
  ID,
  Escolaridade,
  QtdProdutosTotal
FROM
  marketing
WHERE
  Escolaridade IN ('Graduado')
  AND QtdProdutosTotal > 1500
'''
).show(20)
```

```python
# Conta quantos usuários estão em cada categoria de estado_civil

spark.sql(
'''
SELECT
  Estado_Civil,
  COUNT(Estado_Civil) AS Qtd_Estado_Civil
FROM
  marketing
GROUP BY Estado_Civil

'''
).show(10)
```

```python
# Faz uma busca ordenando por Escolaridade, depois por Estado Civil e por Qtd Total de produtos, do maior para o menor

spark.sql(
'''
SELECT
  *
FROM
  marketing
ORDER BY Escolaridade, Estado_Civil, QtdProdutosTotal DESC

'''
).show(20)
```

```python
# Soma as compras da Web e de Loja e mostra em uma nova coluna

spark.sql(
'''
SELECT
  ID,
  NumComprasWeb,
  NumComprasLoja,
  (NumComprasWeb + NumComprasLoja) AS TotalCompras
FROM
  marketing
'''
).show()
```

```python
# Pega todos os clientes desde 2012 que são PhDs e compraram mais de 1500 produtos

spark.sql(
'''
SELECT
  *
FROM
  marketing
WHERE
  Data_Cliente LIKE '2012%'
  AND Escolaridade = 'PhD'
  AND QtdProdutosTotal > 1500
ORDER BY
  QtdProdutosTotal
  
'''
).show(20)  
```

```python
# Seleciona somente os últimos clientes, cadastrados nos últimos três meses de 2014

spark.sql(
'''
SELECT
  ID,
  Data_Cliente
FROM
  marketing
WHERE
  Data_Cliente LIKE '2014-1%'
'''
).show(100)
```

```python
# Achando o cliente mais velho (Parte 1)
spark.sql(
'''
SELECT
  MIN(Ano_Nascimento)
FROM
  marketing

'''
).show()
```
